```python
import pandas as pd

datasets = [1, 2, 3]

for dataset in datasets:
    # Load the data for each dataset
    data_outside_filtered = pd.read_csv(f'Dataset{dataset}/data_outside.csv')
    data_outside_filtered['Irms'] = data_outside_filtered['Irms'].apply(lambda x: x**0.004)
    Irms_avg = data_outside_filtered['Irms'].mean()
    data_outside_filtered['Irms'] = data_outside_filtered['Irms'].apply(lambda x: x - Irms_avg)
    data_outside_filtered['Irms'] = data_outside_filtered['Irms'].apply(lambda x: 0 if x < 0 else x)
    data_outside_filtered['Irms'] = data_outside_filtered['Irms'] * 150
    data_outside_filtered['Watt'] = data_outside_filtered['Irms'] * 230

    data_solar = pd.read_csv(f'Dataset{dataset}/new_solar_data.csv')
    data_solar['time'] = data_solar['date']
    data_solar = data_solar.drop(columns=['date'], axis=1)

    data_outside_filtered['Power (kW)'] = data_outside_filtered['Watt'] / 1000
    data_outside_filtered['time'] = pd.to_datetime(data_outside_filtered['time'])
    data_outside_filtered['Time Difference'] = data_outside_filtered['time'].diff()
    data_outside_filtered['Energy Usage (kWh)'] = data_outside_filtered['Power (kW)'] * data_outside_filtered['Time Difference'].dt.total_seconds() / 3600
    total_energy_usage = data_outside_filtered['Energy Usage (kWh)'].sum()

    solar_df = pd.read_csv(f'Dataset{dataset}/new_solar_data.csv')
    solar_df['Timestamp'] = solar_df['date']
    solar_df = solar_df.drop(columns=['date'], axis=1)
    solar_df.set_index('Timestamp', inplace=True)
    solar_df.index = pd.to_datetime(solar_df.index)

    resampled_solar_df = solar_df.resample('1T').ffill()
    resampled_solar_df = resampled_solar_df.interpolate(method='time')

    data_outside_filtered['Timestamp'] = data_outside_filtered['time']
    merged_df = pd.merge_asof(data_outside_filtered, resampled_solar_df, on='Timestamp', direction='nearest')

    merged_df['Solar Available'] = merged_df['value'].notna()
    merged_df['Solar Available'] = merged_df['value'].apply(lambda x: False if x == 0 else True)

    filtered_df = merged_df[merged_df['Solar Available'] == False]
    energy_usage_without_sun = filtered_df['Energy Usage (kWh)'].sum()

    print(f"Dataset {dataset} Energy Usage Without Solar:", energy_usage_without_sun)
    # calculate total duration of dataset
    total_duration = merged_df['Time Difference'].sum()
    print(f"Dataset {dataset} Total Duration:", total_duration)
    merged_df.to_csv(f'Dataset{dataset}/merged_df.csv')
```

```python
import pandas as pd

datasets = [1, 2, 3]
max_duration = pd.Timedelta(days=0)
total_energy_usage = {}

# Calculate the minimum duration among the datasets
for dataset in datasets:
    merged_df = pd.read_csv(f'Dataset{dataset}/merged_df.csv')
    merged_df['Time Difference'] = pd.to_timedelta(merged_df['Time Difference'])
    total_duration = merged_df['Time Difference'].sum()
    if total_duration > max_duration:
        max_duration = total_duration

# Truncate datasets to have the same duration
for dataset in datasets:
    merged_df = pd.read_csv(f'Dataset{dataset}/merged_df.csv')
    merged_df['Time Difference'] = pd.to_timedelta(merged_df['Time Difference'])
    merged_df['Cumulative Duration'] = merged_df['Time Difference'].cumsum()

    total_duration = merged_df['Time Difference'].sum()
    # Calculate the ratio of truncated duration to original duration
    duration_ratio = total_duration / max_duration

    # Recalculate energy usage based on truncated duration
    filtered_df = merged_df[merged_df['Solar Available'] == False]
    energy_usage_without_sun = filtered_df['Energy Usage (kWh)'].sum() / duration_ratio
    total_energy_usage[dataset] = energy_usage_without_sun

    print(f"Dataset {dataset} Energy Usage Without Solar (Truncated):", energy_usage_without_sun)
    # print(f"Dataset {dataset} Total Duration (Truncated):", total_duration, duration_ratio)
    # calculate usage with solar
    with_solar_df = merged_df[merged_df['Solar Available'] == True]
    usage_during_solar = with_solar_df['Energy Usage (kWh)'].sum() / duration_ratio
    print(f"Dataset {dataset} Energy Usage With Solar (Truncated):", usage_during_solar)



# Print total energy usage based on the truncated duration
for dataset, energy_usage in total_energy_usage.items():
    print(f"Dataset {dataset} Total Energy Usage (Truncated):", energy_usage)
```
